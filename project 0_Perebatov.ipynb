```python
def guess_limit(max_limit, lag_limit=1):
    """Создает список вероятных ограничений для случайного числа кратного делителю."""
    list_limit = []
    limit_1 = max_limit
    limit_2 = max_limit
    for i in range(0, 30):
        if limit_2 == lag_limit :
            list_limit.append(lag_limit)
        else:
            limit_1, limit_2 = limit_2 // 2, limit_1 // 2
            if limit_2 % lag_limit != 0: 
                #проверка кратности
                limit_2 = (limit_1//lag_limit) * lag_limit
            list_limit.append(limit_2)
    return list_limit


def game_guess (number, max_limit=100):
    """Угадывает любое число от 1 - 100 и более за n - попыток"""
    count_guess = 0
    predict = 0
    if   number>max_limit or number<0:
        return print(f'введите число в пределах 0 - {max_limit}')
    
    for lag_limit in range(50,0,-1):
        if number % lag_limit == 0:
            predict_limit = guess_limit(max_limit, lag_limit) 
            # список вида [50,25,10,5,5,5,5,5,5]
            break
        if lag_limit == 1: 
            return print('некорректное число')

    while number != predict:
        count_guess += 1
        if number > predict: 
            predict += predict_limit[count_guess - 1]
        elif number < predict: 
            predict -= predict_limit[count_guess - 1]
    print(f'Вы угадали число с {count_guess} попытки!')    
    return(count_guess)

game_guess(15)
```

```python
import numpy as np

def score_game(game_guess):
    '''Запускаем игру 1000 раз, чтобы узнать, как быстро игра угадывает число'''
    count_ls = []
    np.random.seed(1)  # фиксируем RANDOM SEED, чтобы ваш эксперимент был воспроизводим!
    random_array = np.random.randint(1,101, size=(1000))
    for number in random_array:
        count_ls.append(game_guess(number))
    score = int(np.mean(count_ls))
#     score = count_ls.count(7)
    print(f"Ваш алгоритм угадывает число в среднем за {score} попытки")
    return(score)


score_game(game_guess)
```
